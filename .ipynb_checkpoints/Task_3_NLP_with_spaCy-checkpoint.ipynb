## Amazon Product Reviews Analysis with spaCy

* NLP with Named Entity Recognition and Rule-Based Sentiment Analysis

```python

# library
import spacy
from spacy import displacy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict
import re
import random

# Download the spaCy model if not already installed
# python -m spacy download en_core_web_sm

print("=" * 70)
print("AMAZON PRODUCT REVIEWS ANALYSIS WITH spaCy")
print("=" * 70)

```

```python
# Step 1: Load spaCy model and define custom patterns
print("\n" + "=" * 50)
print("STEP 1: LOADING NLP MODEL AND DEFINING PATTERNS")
print("=" * 50)

# Load the English model
try:
    nlp = spacy.load("en_core_web_sm")
    print("âœ… spaCy model 'en_core_web_sm' loaded successfully!")
except OSError:
    print("âŒ Model not found. Please install with: python -m spacy download en_core_web_sm")
    exit()

# Add custom patterns for product-related entities
print("\nAdding custom entity patterns for product recognition...")

# Define custom patterns for product names and brands
product_patterns = [
    # Product types
    {"label": "PRODUCT", "pattern": [{"LOWER": "iphone"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "macbook"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "ipad"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "airpods"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "kindle"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "echo"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "galaxy"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "playstation"}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "xbox"}]},

    # Product models with numbers
    {"label": "PRODUCT", "pattern": [{"LOWER": "iphone"}, {"IS_DIGIT": True}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "galaxy"}, {"IS_DIGIT": True}]},
    {"label": "PRODUCT", "pattern": [{"LOWER": "series"}, {"IS_DIGIT": True}]},
]

# Add patterns to the entity ruler
ruler = nlp.add_pipe("entity_ruler", before="ner")
ruler.add_patterns(product_patterns)

print("âœ… Custom entity patterns added successfully!")

```

```python
# Step 2: Define sentiment analysis components
print("\n" + "=" * 50)
print("STEP 2: SETTING UP RULE-BASED SENTIMENT ANALYSIS")
print("=" * 50)

# Define sentiment lexicons
positive_words = {
    'excellent', 'amazing', 'great', 'good', 'awesome', 'fantastic', 'perfect',
    'outstanding', 'superb', 'wonderful', 'brilliant', 'love', 'loved', 'best',
    'recommend', 'happy', 'satisfied', 'pleased', 'impressive', 'worth',
    'quality', 'reliable', 'durable', 'fast', 'easy', 'smooth', 'comfortable',
    'beautiful', 'stunning', 'exceeded', 'outstanding', 'flawless'
}

negative_words = {
    'terrible', 'awful', 'horrible', 'bad', 'poor', 'disappointing', 'disappointed',
    'waste', 'rubbish', 'garbage', 'useless', 'broken', 'defective', 'faulty',
    'slow', 'difficult', 'hard', 'uncomfortable', 'ugly', 'cheap', 'overpriced',
    'regret', 'return', 'refund', 'complaint', 'issue', 'problem', 'error',
    'failure', 'damaged', 'missing', 'wrong'
}

# Intensifiers that amplify sentiment
intensifiers = {
    'very', 'extremely', 'really', 'absolutely', 'completely', 'totally',
    'utterly', 'highly', 'particularly', 'especially'
}

# Negation words that flip sentiment
negation_words = {
    'not', "n't", 'no', 'never', 'nothing', 'none', 'neither', 'nor'
}

print("âœ… Sentiment lexicons defined:")
print(f"   - Positive words: {len(positive_words)}")
print(f"   - Negative words: {len(negative_words)}")
print(f"   - Intensifiers: {len(intensifiers)}")
print(f"   - Negation words: {len(negation_words)}")

def analyze_sentiment(text):
    """
    Rule-based sentiment analysis for product reviews
    Returns: sentiment label and score
    """
    doc = nlp(text.lower())
    tokens = [token.text for token in doc]

    positive_score = 0
    negative_score = 0

    # Check for sentiment words with context
    for i, token in enumerate(doc):
        word = token.text

        # Check if current word is a sentiment word
        sentiment_strength = 1.0
        is_negated = False
        is_intensified = False

        # Check for negation in previous words
        for j in range(max(0, i-3), i):
            if doc[j].text in negation_words:
                is_negated = True
                break

        # Check for intensifiers in previous words
        for j in range(max(0, i-2), i):
            if doc[j].text in intensifiers:
                is_intensified = True
                sentiment_strength *= 2.0  # Double the strength
                break

        # Apply sentiment scoring
        if word in positive_words:
            if is_negated:
                negative_score += sentiment_strength
            else:
                positive_score += sentiment_strength

        elif word in negative_words:
            if is_negated:
                positive_score += sentiment_strength
            else:
                negative_score += sentiment_strength

    # Calculate overall sentiment
    total_score = positive_score - negative_score

    if total_score > 1:
        return "positive", total_score
    elif total_score < -1:
        return "negative", total_score
    else:
        return "neutral", total_score

print("âœ… Sentiment analysis function defined!")

```

```python
# Step 3: Sample Amazon product reviews data
print("\n" + "=" * 50)
print("STEP 3: SAMPLE AMAZON PRODUCT REVIEWS DATA")
print("=" * 50)

# Sample Amazon product reviews
sample_reviews = [
    {
        "review": "I absolutely love my new iPhone 14! The camera quality is outstanding and the battery life exceeds my expectations. Apple has done an amazing job with this product.",
        "rating": 5
    },
    {
        "review": "The Samsung Galaxy S23 is a terrible phone. It constantly freezes and the battery drains too quickly. I regret buying this garbage.",
        "rating": 1
    },
    {
        "review": "My MacBook Pro is very reliable for work. The performance is good but the price was a bit high. Overall, I'm satisfied with this laptop from Apple.",
        "rating": 4
    },
    {
        "review": "The Sony headphones I bought are not very comfortable. The sound quality is poor and they broke after two weeks. Definitely not worth the money.",
        "rating": 2
    },
    {
        "review": "Amazing product! The Kindle Paperwhite has completely changed my reading habits. The display is perfect and battery lasts for weeks. Highly recommend!",
        "rating": 5
    },
    {
        "review": "The Microsoft Surface laptop is okay. Nothing special but it gets the job done. The build quality could be better for the price they're asking.",
        "rating": 3
    },
    {
        "review": "I hate this HP printer. It's constantly having connectivity issues and the ink cartridges are overpriced. Worst purchase ever from Amazon.",
        "rating": 1
    },
    {
        "review": "The Dell monitor is fantastic! The colors are vibrant and the setup was extremely easy. Great value for money and very pleased with this purchase.",
        "rating": 5
    }
]

print("Sample reviews loaded:")
for i, review_data in enumerate(sample_reviews, 1):
    print(f"\nReview {i} (Rating: {review_data['rating']}/5):")
    print(f"  {review_data['review']}")
```

```python

# Step 4: Process reviews and extract entities
print("\n" + "=" * 50)
print("STEP 4: PROCESSING REVIEWS & EXTRACTING ENTITIES")
print("=" * 50)

results = []

for i, review_data in enumerate(sample_reviews, 1):
    review_text = review_data["review"]
    rating = review_data["rating"]

    print(f"\nðŸ“ Processing Review {i} (Rating: {rating}/5):")
    print(f"   Text: {review_text}")

    # Process with spaCy
    doc = nlp(review_text)

    # Extract entities
    entities = []
    for ent in doc.ents:
        if ent.label_ in ["ORG", "PRODUCT", "GPE"]:  # Organization, Product, Geographical Entity
            entities.append({
                "text": ent.text,
                "label": ent.label_,
                "start": ent.start_char,
                "end": ent.end_char
            })

    # Analyze sentiment
    sentiment, score = analyze_sentiment(review_text)

    # Store results
    result = {
        "review_id": i,
        "text": review_text,
        "rating": rating,
        "entities": entities,
        "sentiment": sentiment,
        "sentiment_score": score
    }
    results.append(result)

    # Print extracted entities
    if entities:
        print("   ðŸ” Extracted Entities:")
        for entity in entities:
            print(f"      - {entity['text']} ({entity['label']})")
    else:
        print("   ðŸ” No relevant entities found")

    print(f"   ðŸ˜Š Sentiment: {sentiment} (score: {score:.2f})")

```

```python
# Step 5: Comprehensive Analysis and Visualization
print("\n" + "=" * 50)
print("STEP 5: COMPREHENSIVE ANALYSIS & VISUALIZATION")
print("=" * 50)

# Convert results to DataFrame for analysis
df_results = pd.DataFrame(results)

# Entity Analysis
print("\nðŸ“Š ENTITY ANALYSIS:")
all_entities = []
for result in results:
    for entity in result["entities"]:
        all_entities.append((entity["text"], entity["label"]))

entity_counter = Counter(all_entities)
print("Most common entities:")
for (entity, label), count in entity_counter.most_common(10):
    print(f"   {entity} ({label}): {count} occurrences")

# Sentiment Analysis
print("\nðŸ“Š SENTIMENT ANALYSIS SUMMARY:")
sentiment_counts = df_results["sentiment"].value_counts()
print(sentiment_counts)

# Compare with actual ratings
print("\nðŸ“ˆ SENTIMENT vs RATING COMPARISON:")
for sentiment in ["positive", "neutral", "negative"]:
    sentiment_data = df_results[df_results["sentiment"] == sentiment]
    avg_rating = sentiment_data["rating"].mean()
    print(f"   {sentiment.capitalize()} sentiment: Average rating = {avg_rating:.2f}/5")

```

```python
# Step 6: Visualize the results
print("\n" + "=" * 50)
print("STEP 6: CREATING VISUALIZATIONS")
print("=" * 50)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Sentiment Distribution
sentiment_counts.plot(kind='bar', ax=axes[0, 0], color=['green', 'gray', 'red'])
axes[0, 0].set_title('Sentiment Distribution in Reviews')
axes[0, 0].set_xlabel('Sentiment')
axes[0, 0].set_ylabel('Number of Reviews')

# 2. Entity Type Distribution
entity_types = [entity[1] for entity in all_entities]
type_counter = Counter(entity_types)
pd.Series(dict(type_counter)).plot(kind='bar', ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Entity Types Extracted')
axes[0, 1].set_xlabel('Entity Type')
axes[0, 1].set_ylabel('Count')

# 3. Sentiment Score Distribution
df_results['sentiment_score'].plot(kind='hist', ax=axes[1, 0], bins=10, color='purple', alpha=0.7)
axes[1, 0].set_title('Distribution of Sentiment Scores')
axes[1, 0].set_xlabel('Sentiment Score')
axes[1, 0].set_ylabel('Frequency')

# 4. Rating vs Sentiment Score
axes[1, 1].scatter(df_results['rating'], df_results['sentiment_score'], alpha=0.7, s=100)
axes[1, 1].set_title('Rating vs Sentiment Score')
axes[1, 1].set_xlabel('Star Rating (1-5)')
axes[1, 1].set_ylabel('Sentiment Score')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
# Step 7: Detailed Entity-Sentiment Analysis
print("\n" + "=" * 50)
print("STEP 7: ENTITY-SENTIMENT CORRELATION")
print("=" * 50)

# Analyze sentiment around specific entities
entity_sentiment = defaultdict(list)

for result in results:
    doc = nlp(result["text"])
    sentiment = result["sentiment"]

    for ent in doc.ents:
        if ent.label_ in ["ORG", "PRODUCT"]:
            entity_sentiment[ent.text].append(sentiment)

print("Entity Sentiment Analysis:")
for entity, sentiments in entity_sentiment.items():
    positive_count = sentiments.count("positive")
    negative_count = sentiments.count("negative")
    neutral_count = sentiments.count("neutral")
    total = len(sentiments)

    print(f"\nðŸ·ï¸  {entity}:")
    print(f"   Positive: {positive_count}/{total} ({positive_count/total*100:.1f}%)")
    print(f"   Negative: {negative_count}/{total} ({negative_count/total*100:.1f}%)")
    print(f"   Neutral:  {neutral_count}/{total} ({neutral_count/total*100:.1f}%)")

```

```python
# Step 8: Interactive Entity Visualization
print("\n" + "=" * 50)
print("STEP 8: INTERACTIVE ENTITY VISUALIZATION")
print("=" * 50)

# Select a sample review for detailed visualization
sample_review = sample_reviews[0]["review"]
doc = nlp(sample_review)

print("Sample Review with Entities Highlighted:")
print(f"Review: {sample_review}")
print("\nEntities found:")
colors = {"ORG": "lightblue", "PRODUCT": "lightgreen", "GPE": "yellow"}
for ent in doc.ents:
    if ent.label_ in colors:
        print(f"  - '{ent.text}' ({ent.label_})")

# Generate HTML visualization
# html = displacy.render(doc, style="ent", options={"colors": colors})
# with open("entity_visualization.html", "w", encoding="utf-8") as f:
#      f.write(html)
# print("âœ… Entity visualization saved as 'entity_visualization.html'")

```

```python
# Step 9: Performance Evaluation
print("\n" + "=" * 50)
print("STEP 9: PERFORMANCE EVALUATION")
print("=" * 50)

# Compare rule-based sentiment with actual ratings
correct_predictions = 0
total_predictions = len(df_results)

for _, row in df_results.iterrows():
    actual_rating = row['rating']
    predicted_sentiment = row['sentiment']

    # Map ratings to sentiment
    if actual_rating >= 4:
        actual_sentiment = "positive"
    elif actual_rating <= 2:
        actual_sentiment = "negative"
    else:
        actual_sentiment = "neutral"

    if predicted_sentiment == actual_sentiment:
        correct_predictions += 1

accuracy = correct_predictions / total_predictions

print(f"Rule-Based Sentiment Analysis Performance:")
print(f"âœ… Accuracy: {accuracy:.2%} ({correct_predictions}/{total_predictions} correct)")
print(f"ðŸ“Š This measures how well our rule-based approach matches the actual star ratings")

```
